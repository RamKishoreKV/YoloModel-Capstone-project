```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from collections import Counter
from sklearn.model_selection import KFold
from PIL import Image
```

```python
dataset_dirs = {
    "train": {
        "image_dir": "datasets/train/images",  # Path to train images
        "label_dir": "datasets/train/labels"     # Path to train labels
    },
    "test": {
        "image_dir": "datasets/val/images",     # Path to test images
        "label_dir": "datasets/val/labels"        # Path to test labels
    }
}
```

```python
split_dir = "datasets/splits"
os.makedirs(split_dir, exist_ok=True)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
train_images = [f for f in os.listdir(dataset_dirs["train"]["image_dir"]) if f.endswith((".jpg", ".png", ".jpeg"))]

for fold, (train_idx, val_idx) in enumerate(kf.split(train_images), 1):
    train_files = [train_images[i] for i in train_idx]
    val_files = [train_images[i] for i in val_idx]

    with open(os.path.join(split_dir, f"train_fold{fold}.txt"), "w") as f:
        f.writelines("\n".join(train_files))
    
    with open(os.path.join(split_dir, f"val_fold{fold}.txt"), "w") as f:
        f.writelines("\n".join(val_files))

    print(f"Fold {fold}: Train={len(train_files)}, Validation={len(val_files)}")

print("5-Fold Cross-Validation splits saved.")
```

```python
import os

def check_labels_in_file(file_path):
    """
    Checks if the file contains at least one occurrence of '0' or '1'.

    :param file_path: Path to the text file.
    :return: True if '0' or '1' is found, otherwise False.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            return '0' in content or '1' in content
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return False

def count_label_files(directory):
    """
    Counts the number of text files in the given directory that contain '0' or '1'.

    :param directory: Path to the directory containing text files.
    :return: The count of files that contain '0' or '1'.
    """
    count = 0
    if os.path.exists(directory) and os.path.isdir(directory):
        for file in os.listdir(directory):
            if file.endswith('.txt'):
                file_path = os.path.join(directory, file)
                if check_labels_in_file(file_path):
                    count += 1
    return count

# Define the paths
train_labels_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/train/labels"
val_labels_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/test/labels"

# Get counts of valid text files in label folders
train_valid_count = count_label_files(train_labels_path)
val_valid_count = count_label_files(val_labels_path)

# Print the results
print(f"Number of text files in train/labels containing '0' or '1': {train_valid_count}")
print(f"Number of text files in test/labels containing '0' or '1': {val_valid_count}")
```

```python
import os

def check_labels_in_file(file_path):
    """
    Checks if the file contains at least one occurrence of '0' or '1'.

    :param file_path: Path to the text file.
    :return: True if '0' or '1' is found, otherwise False.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            return '0' in content or '1' in content
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return False

def clean_labels_and_images(labels_dir, images_dir):
    """
    Deletes text files that do not contain '0' or '1' and their corresponding image files.

    :param labels_dir: Path to the directory containing label text files.
    :param images_dir: Path to the directory containing corresponding images.
    """
    if not os.path.exists(labels_dir) or not os.path.isdir(labels_dir):
        print(f"Label directory not found: {labels_dir}")
        return

    if not os.path.exists(images_dir) or not os.path.isdir(images_dir):
        print(f"Image directory not found: {images_dir}")
        return

    for file in os.listdir(labels_dir):
        if file.endswith('.txt'):
            file_path = os.path.join(labels_dir, file)
            image_file = file.replace('.txt', '.jpg')
            image_path = os.path.join(images_dir, image_file)

            # If the label file does not contain '0' or '1', delete it and the corresponding image
            if not check_labels_in_file(file_path):
                try:
                    os.remove(file_path)  # Delete text file
                    print(f"Deleted label file: {file_path}")

                    if os.path.exists(image_path):
                        os.remove(image_path)  # Delete corresponding image
                        print(f"Deleted image file: {image_path}")

                except Exception as e:
                    print(f"Error deleting files: {e}")

# Define the paths
train_labels_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/train/labels"
train_images_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/train/images"
val_labels_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/test/labels"
val_images_path = "/Users/adityasinha/Desktop/Drexel/Capstone1/datasets/test/images"

# Clean the train and test label & image folders
clean_labels_and_images(train_labels_path, train_images_path)
clean_labels_and_images(val_labels_path, val_images_path)

print("Cleanup completed!")
```
